# Polynomial Linear Regression

<img src="polynomialRegression.png" alt="Polynomial Regression" style="width: 700px;"/>

Polynomial linear regression is the concept of determining a polynomial relationship between an independent and dependent variable. The regression is known as linear regression because only one independent variable is utilized. The code functions very similar to multivariate linear regression, because the single independent variable is split into many columns based on the power of x.

__Before Tranformation:__

| X |
|---|
| 1 |
| 2 |
| 3 |
| 4 |
| 5 |

__After Transformation:__

| $X^0$ | $X^1$ | $X^2$ | $X^3$ |
|-------|-------|-------|-------|
| 1     | 1     | 1     | 1     |
| 1     | 2     | 4     | 8     |
| 1     | 3     | 9     | 27    |
| 1     | 4     | 16    | 64    |
| 1     | 5     | 25    | 125   |

## Code

__Setting up the dataset:__

```python
# Data Preprocessing Template

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

<hr>

__Performing Linear Regression:__

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Graph Scaling
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 10
fig_size[1] = 7.5

# Visual Representation (Linear Regression)
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.ylabel('Position Level')
plt.xlabel('Salary')
plt.show()
```

<hr>

__Performing Polynomial Regression:__

```python
# Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4) #Higher Degree = Higher Accuracy
X_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)

#Visual Representation (Polynomial Regression)
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.ylabel('Position Level')
plt.xlabel('Salary')
plt.show()
```

<hr>

__Smoothing the Graph:__

```python
#Visual Representation (Polynomial Regression)
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)

plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.ylabel('Position Level')
plt.xlabel('Salary')
plt.show()
```

__Predicting a Singular Value:__

```python
lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```
